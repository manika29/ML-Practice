# Content Based Filtering

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read csv file
```

```python
movies=pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
def get_index_from_title(title):
    return movies[movies.title == title]["movieId"].values[0]
```

```python
def get_title_from_index(index):
    return movies[movies.index == index]["title"].values[0]
```

```python
# using genres as a feature to recommend movies
features=['genres']
```

```python
for feature in features:
    movies[feature] = movies[feature].fillna('')
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(movies["genres"])
```

```python
# Compute the Cosine Similarity based on the count_matrix
```

```python
cosine_sim=cosine_similarity(count_matrix)
```

```python
cosine_sim
```

```python
# Recommend Movies
```

```python
movie_user_likes="Toy Story (1995)"
```

```python
# Get index of this movie from its title
```

```python
movie_index = get_index_from_title(movie_user_likes)
```

```python
similar_movies =  list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
i=0
list=[]
for element in sorted_similar_movies:
        movie = (get_title_from_index(element[0]))
        list.append(movie)
        i=i+1
        if i>5:
            break
```

```python
movie_recommendation1 = pd.DataFrame(list,columns=['MOVIES'])
```

# Collaborative Filtering

```python
# Read movies and ratings csv files
```

```python
ratings=pd.read_csv('ratings.csv')
movies=pd.read_csv('movies.csv')
ratings=pd.merge(movies,ratings).drop(['timestamp'],axis=1)
ratings.head()
```

```python
user_ratings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
# Remove movies that have less than 10 users who rated it and fill remaining nan with 0.
user_ratings=user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings
```

```python
# Let's build our similarity matrix
corrMatrix = user_ratings.corr(method='pearson')
corrMatrix.head(5)
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
action_movies_user_likes = [("Amazing Spider-Man, The (2012)",5),("Mission: Impossible III (2006)",4),("Toy Story 3 (2010)",2),("2 Fast 2 Furious (Fast and the Furious 2, The) (2003)",4)]
similar_movies = pd.DataFrame()
for movie,rating in action_movies_user_likes:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)

similar_movies.head(10)
movie_recommendation2=similar_movies.sum().sort_values(ascending=False).head(5)
```
