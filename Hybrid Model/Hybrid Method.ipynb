# Hybrid Recommendation System

1 - CONTENT BASED RECOMMENDATION

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read csv file
```

```python
movie_df=pd.read_csv('movie_dataset.csv')
```

```python
movie_df.head()
```

```python
# Select features
```

```python
features=['keywords','cast','genres','director']
```

```python
# Fill all the NaN values
```

```python
for feature in features:
    movie_df[feature] = movie_df[feature].fillna('')
```

```python
# Combine all these features in one column
```

```python
def combine_features(row):
    try:
        return row['keywords'] +" "+row['cast']+" "+row["genres"]+" "+row["director"]
    except:
        print ("Error:", row)
```

```python
movie_df["combined_features"] = movie_df.apply(combine_features,axis=1)
```

```python
# Create count matrix from this new combined column
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(movie_df["combined_features"])
```

```python
# Compute the Cosine Similarity based on the count_matrix
```

```python
cosine_sim=cosine_similarity(count_matrix)
cosine_sim
```

```python
# Recommend Movies
```

```python
movie_user_likes="Toy Story 3"
```

```python
# Get index of this movie from its title
```

```python
def get_index_from_title(title):
    return movie_df[movie_df.title == title]["index"].values[0]
```

```python
movie_index = get_index_from_title(movie_user_likes)
```

```python
similar_movies =  list(enumerate(cosine_sim[movie_index]))
```

```python
# Get a list of similar movies in descending order of similarity score
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
# Print titles of first 5 movies
```

```python
def get_title_from_index(index):
    return movie_df[movie_df.index == index]["title"].values[0]
```

```python
i=0
for element in sorted_similar_movies:
        print(get_title_from_index(element[0]))
        i=i+1
        if i>5:
            break
```

2 - COLLABORATIVE FILTERING BASED RECOMMENDATION

```python
# Read csv
```

```python
ratings=pd.read_csv('ratings.csv')
movies=pd.read_csv('movie_dataset.csv')
ratings=pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
ratings.head()
```

```python
user_ratings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
# Remove movies that have less than 10 users who rated it and fill remaining nan with 0.
user_ratings=user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings
```

```python
# Let's build our similarity matrix
corrMatrix = user_ratings.corr(method='pearson')
corrMatrix.head(5)
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
user_movies = [("Toy Story 3 (2010)",2)]
similar_movies = pd.DataFrame()
for movie,rating in user_movies:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)

similar_movies.head(10)
similar_movies.sum().sort_values(ascending=False).head(5)
```
